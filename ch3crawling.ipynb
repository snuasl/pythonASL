```python
import urllib.request
from bs4 import BeautifulSoup as BS

mat_id = "C7727379" # Liquid Nitrogen 
P = 1e5 # const Pressure [Pa]
T2 = 300 # Temp [K]
T1 = 77 # Temp [K]
Tinc = 0.1 # Temp increment [K]

# Isobaric transition data 
link = "https://webbook.nist.gov/cgi/fluid.cgi?"
link = "https://webbook.nist.gov/cgi/fluid.cgi?"
link = link + "Action=Data&Wide=on&ID=%s&" %(mat_id)
link = link + "Type=IsoBar&Digits=5&P=%f&" %(P/1e6)
link = link + "THigh=%d&TLow=%d&TInc=%d&" %(T2, T1, Tinc)
link = link + "RefState=DEF&TUnit=K&PUnit=MPa&"
link = link + "DUnit=kg"+"%"+"2Fm3&HUnit=kJ"+"%"+"2Fkg&WUnit=m"+"%"+"2Fs&"
link = link + "VisUnit=Pa*s&STUnit=N"+"%"+"2Fm"

page = urllib.request.urlopen(link).read()
soup = BS(page, features="lxml")
body = soup.find('body')
data = body.findChildren(recursive=False)
data = body.get_text()

print(data)
```
